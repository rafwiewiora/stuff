```python
import keras
from keras.models import model_from_json
from keras.preprocessing.sequence import pad_sequences
import numpy as np
from utils import lettercode2onehot
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
with open('model.json') as f:
    model = model_from_json(f.read())
```

```python
model.load_weights('model.h5')
```

```python
# predict WT M. smeg. peptide
```

```python
reca_peptide = np.array(['N', 'Q', 'L', 'R', 'E', 'K', 'I', 'G', 'V', 'M', 'F', 'G', 'S', 'P', 'E', 'T', 'T', 'T', 'G', 'G'])
```

```python
X = [lettercode2onehot(reca_peptide)]
X = pad_sequences(X, 20*800)
```

```python
Y = model.predict(X)
```

```python
# 'unpad' prediction
reca_peptide_prediction = Y[0][-(len(reca_peptide)):]
```

```python
fig = plt.figure(dpi=300)
ax = fig.add_subplot(1,1,1)
ax.bar((np.arange(len(reca_peptide))), reca_peptide_prediction)
ax.set_xticks(np.arange(len(reca_peptide)))
ax.set_xticklabels(reca_peptide)
ax.set_xlabel('Residue')
ax.set_ylabel('Structural propensity')
plt.show() 
```
